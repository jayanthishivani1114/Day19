```python
import subprocess
from pathlib import Path

def run(cmd):
    print(f"\n$ {cmd}")
    result = subprocess.run(cmd, shell=True, text=True, capture_output=True)
    print(result.stdout)
    if result.stderr:
        print(result.stderr)

if not Path(".git").exists():
    run("git init")
    run("git branch -M main")

run("git checkout main")

file_name = "conflict_demo.py"
if not Path(file_name).exists():
    with open(file_name, "w") as f:
        f.write("print('Version from MAIN')\n")
    run(f"git add {file_name}")
    run('git commit -m "Initial commit on main"')

with open(file_name, "w") as f:
    f.write("print('Edited in MAIN branch')\n")

run(f"git add {file_name}")
run('git commit -m "Edit line 1 on main"')

run("git checkout -b feature-conflict")

with open(file_name, "w") as f:
    f.write("print('Edited in FEATURE branch')\n")

run(f"git add {file_name}")
run('git commit -m "Edit line 1 on feature branch"')
# Switch back to main
run("git checkout main")

# Try merging (this will create a conflict)
run("git merge feature-conflict")

print("\nNow open the file to see conflict markers:\n")
with open(file_name, "r") as f:
    print(f.read())

print("""
Manual Fix Steps:
1. Open the file.
2. Find the conflict markers:
   <<<<<<< HEAD
   =======
   >>>>>>> feature-conflict
3. Choose the version you want (or combine them).
4. Remove ALL conflict markers.
5. Save the file.
6. Run:

   git add conflict_demo.py
   git commit -m "Resolve merge conflict"

Merge complete.
""")
```
